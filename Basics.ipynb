```python
## line continuation - You can use a backslash (\) to indicate that a statement continues on the next line.
total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)
```

```python
## Multiple statements on a single line - You can use a semicolon (;) to separate multiple statements on a single line.

x = 5; y = 10; z = 15   

print(x, y, z)
print(x)
print(y)
print(z)
```

```python
## Understand Semantics in Python 
## variable assignment - In Python, you can assign values to variables using the assignment operator (=). For example, x = 5 assigns the value 5 to the variable x. 

age = 32 ## age is a integer variable
name = "Krish" ## name is a string variable
print(age)
print(name)
print(type(age))
print(type(name))
```
